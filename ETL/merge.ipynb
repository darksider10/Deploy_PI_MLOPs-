```python
import pandas as pd
```

```python
games = pd.read_csv("games.csv")
items = pd.read_csv("user_items.csv")
reviews = pd.read_csv("user_reviews.csv")
```

```python
# Selecciona solo las columnas 'user_id', 'item_id', 'playtime_forever' del DataFrame 'items'.
items = items[['user_id', 'item_id', 'playtime_forever']]

# Convierte la columna a tipo cadena.
items['item_id'] = items['item_id'].astype(str)

# Crea una nueva columna 'id' concatenando 'user_id' y 'item_id'.
items['id'] = items['user_id'] + items['item_id']

# Convierte las columnas a tipo cadena.
games['item_id'] = games['item_id'].astype(str)
reviews['item_id'] = reviews['item_id'].astype(str)

# Crea una nueva columna 'id' en el DataFrame 'reviews' concatenando 'user_id' y 'item_id'.
reviews['id'] = reviews['user_id'] + reviews['item_id']
```

```python
# Realiza un merge entre los DataFrames 'reviews' y 'games' basándose en la columna 'item_id'.
merged_df = reviews.merge(games, on='item_id', how='left')

# Elimina filas con valores nulos en el DataFrame resultante.
merged_df.dropna(inplace=True)

# Realiza un merge entre los DataFrames 'items' y 'merged_df' basándose en la columna 'id'.
steam = items.merge(merged_df, on='id')

# Renombra las columnas 'user_id_x' y 'item_id_x' en 'steam'.
steam = steam.rename(columns={'user_id_x': 'user_id', 'item_id_x': 'item_id'})

# Elimina las columnas 'user_id_y' e 'item_id_y' en 'steam'.
steam.drop('user_id_y', axis='columns', inplace=True)
steam.drop('item_id_y', axis='columns', inplace=True)
```

```python
# Convierte las columnas a tipo entero.
steam['release_date'] = steam['release_date'].astype(str)
steam['year_posted'] = steam['year_posted'].astype(str)
steam['playtime_forever'] = steam['playtime_forever'].astype(int)
steam["release_date"] = steam["release_date"].astype(int)

# Selecciona las columnas específicas en el DataFrame 'steam'.
steam = steam[['id', 'user_id', 'item_id', 'title', 'genre', 'developer', 'release_date', 'price', 'recommend', 'year_posted', 'sentiment_analysis', 'playtime_forever']]


steam.to_csv('data_steam.csv', index=False)
```
